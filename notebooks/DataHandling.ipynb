<h1>Interpolation</h1>

```python
series = pd.Series([0,None,5,None,None,35])
ser_linear = series.interpolate()
ser_quadratic = series.interpolate(method='quadratic')

plt.plot([i for i in range(6)],ser_linear)

print('''Linear Interpolating:''')
print('first value is 2.5 obtained by linear interpolating between 0 and 5')
print('the second,third null value is obtained by linear interpolating between 5 and 35 (15,25)')
```
